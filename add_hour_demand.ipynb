```python
import pandas as pd
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")


df = pd.read_csv('rides_final.csv', sep=',', skipinitialspace=True, encoding='utf-8')
```

```python
df.columns
```

```python
index = pd.to_datetime(df['start_date'].values, format='%Y-%m-%d %H:%M:%S')
series = pd.Series(1, index=index)
hour_demand = series.groupby(series.index.hour).sum()
hour_demand
```

```python
plt.title('Спрос на самокаты по часам')
sns.barplot(hour_demand)
```
